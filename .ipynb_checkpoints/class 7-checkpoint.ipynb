# nested for loop

```python
sur_name = ("Syed","M.","Mr")
name= ("Moiz","Ammar","Hassan")

for surname in sur_name:
    for names in name:
        print (surname,names)
    print ("-*"*30)
```

```python
sur_name = ("Syed","M.","Mr")
name= ("Moiz","Ammar","Hassan")
suffix = ("(Engr.)","(Data Analyst)","(Flutter developer)")

for surname in sur_name:
    for names in name:
        for suffixes in suffix:
            print (surname,names,suffixes)
        # print (surname,names)
    print (" * "*30)
```

```python
child = ("Hashim","Naqash","Aneeq")
nationality = ("Pakistani","Nigerian","American")

for children in child:
    for nation in nationality:
        print(children, nation)
    print(" * "*30)
```

```python
# is dictionary mn values  print hngi

dic = {
    "Name":"Abdul",
    "Email":"moizpc2021@gmail.com",
    "Education":"Graduation"
}

for keys, values in dic.items():    #dic.items() issy dictionary li keys k sath values b print hngi
    print(f'{keys} : {values}')
    
```

# While loop

```python
employees = []
count = ""

while count != "exit":
    admin = input("Enter Employee name...:")
    if admin == "exit":
       count = "exit"
    else: 
        employees.append(admin)
print(employees)
```
